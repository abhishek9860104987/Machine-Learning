```python
import matplotlib.pyplot as plt

data['Fuel_Type'].value_counts().plot(kind='bar', color='salmon')
plt.title('Fuel Type Count')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage')
plt.show()
```

```python
import matplotlib.pyplot as plt


brand_counts = data['Brand'].value_counts().head(10)  

plt.figure(figsize=(8, 8))
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Car Brands Distribution')
plt.axis('equal')
plt.show()
```
